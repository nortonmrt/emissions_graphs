# Libraries

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

# Data import - 1st set of graphs

```python
# alimentações: 34, 114, 194, 274

# configurações dos gráficos (fonte e tamanho)
matplotlib.rc('font', family = 'Times New Roman')
matplotlib.rc('font', size = 14)

# leitura dos dados da sonda de gases Testo
data = pd.read_excel("C:/Users/Cliente Especial/Desktop/Engenharia/2022.1/Documentos TCC/Emissões temp phi 0,7/3Phi0,7/emissoes3.xlsx")
temperature = pd.read_excel("C:/Users/Cliente Especial/Desktop/Engenharia/2022.1/Documentos TCC/Emissões temp phi 0,7/3Phi0,7/tempo3.xlsx")

# variáveis de armazenamento dos dados
seg = list(data['seg'])
ppm_CO = list(data['ppm CO'])
ppm_NOx = list(data['ppm NOx'])
ppm_CxHy = list(data['ppm CxHy'])
percentage_CO2 = list(data['% CO2IR'])
percentage_O2 = list(data['% O2'])
temperatures = list(temperature['temperature'])
```

# 1st set of graphs

```python
fig, (ax, ax2, ax3) = plt.subplots(3,1, figsize=(15,20))
#fig.suptitle('Emissões')

# graph CO2/O2 vs time
# eixo x compartilhado
ax_c = ax.twinx()

# Linha 1: emissões de CO
line1, = ax.plot(seg, ppm_CO, 'r', dashes=[8, 4],label = 'CO')
# Linha 2: emissões de NOx
line2, = ax_c.plot(seg, ppm_NOx, dashes=[1, 2],label = 'NOx')

# Linhas representando as alimentações
ax.plot([34,34],[0,600],'b',dashes=[10, 10])
ax.plot([114,114],[0,600],'b',dashes=[10, 10])
ax.plot([194,194],[0,600],'b',dashes=[10, 10])
ax.plot([274,274],[0,600],'b',dashes=[10, 10])

# legendas
ax.legend(loc='lower left')
ax_c.legend(loc='lower right')
ax_c.set_ylim([0,150])
#ax.set_ylim([0,2800])
ax.set(xlabel='Tempo (s)', ylabel='Concentração de CO (ppm)', title='Concentrações de NOx e CO')
ax_c.set(ylabel='Concentração NOx (ppm)')

# graph CO2/O2 vs time
# Eixo x compartilhado
ax_d = ax2.twinx()

# Linha 4: emissões de CO2
line4, = ax2.plot(seg, percentage_CO2, dashes=[1,2], label = '%CO2' )
# Linha 5: emissões de O2
line5, = ax_d.plot(seg, percentage_O2,'r', dashes=[4,8], label = '%O2' )

# Linhas representando as alimentações
ax2.plot([34,34],[1.2,2.6],'b',dashes=[10, 10])
ax2.plot([114,114],[1.2,2.6],'b',dashes=[10, 10])
ax2.plot([194,194],[1.2,2.6],'b',dashes=[10, 10])
ax2.plot([274,274],[1.2,2.6],'b',dashes=[10, 10])

# Legendas
ax2.set(xlabel='Tempo (s)', ylabel='Concentração de CO2 (%)', title='Concentrações de CO2 e O2')
ax_d.set(ylabel='Concentração de O2 (%)')
ax_d.legend(loc='lower right')
ax2.legend(loc='lower left')
#ax2.grid()

# graph temperature/CxHy vs time
# Eixo x compartilhado
ax_e = ax3.twinx()

# Linha 6: temperatura
line6, = ax3.plot(seg, temperatures, 'r', label = 'Temperatura' )
# Linha 7: CxHy
line7, = ax_e.plot(seg, ppm_CxHy, 'g', dashes=[4, 3],label = 'Hidrocarbonetos')

# Linhas representando as alimentações
ax3.plot([34,34],[570,630],'b',dashes=[10, 10])
ax3.plot([114,114],[570,630],'b',dashes=[10, 10])
ax3.plot([194,194],[570,630],'b',dashes=[10, 10])
ax3.plot([274,274],[570,630],'b',dashes=[10, 10])

# legendas
ax3.set(xlabel='Tempo (s)', ylabel='Temperatura (°C)', title='Temperatura e concentração de hidrocarbonetos')
ax_e.set(ylabel='Concentração de hidrocarbonetos (ppm)')
ax_e.set_ylim([2300,2600])
ax_e.legend(loc = 'lower right')
ax3.legend(loc = 'lower left')

#fig.savefig("test.png")
plt.show()
```
